```python
# Generator ex1
def generator_ex1():
    print("Start")
    yield "A point"
    print("Continue")
    yield "B Point"
    print("End")
```

```python
temp = iter(generator_ex1())
```

```python
for i in generator_ex1():
    print(i)
```

```python
temp2 = [x * 3 for x in generator_ex1()]
temp3 = (x * 3 for x in generator_ex1())

print(temp2)
print(temp3)
```

```python
for i in temp2:
    print(i)
print("-"*10)
for i in temp3:
    print(i)
```

* 여기서 생기는 의문점 튜플은 함수 내부에 프린트문이 작동을 하지만 리스트는 작동을 하지 않는다??

>  temp2는 리스트 컴프리헨션을 사용하여 즉시 실행이 된다. generator_ex1() 함수가 즉시 실행되고, 결과가 리스트로 저장이 된다. 따라서, for 루프를 사용할때 내부의 메시기자 출력되지 않는다.

> temp3는 제너레이터 컴프리헨션을 사용하여 지연 평가를 한다. generator_ex1() 함수는 for 루프를 사용하여 반복(iterate) 할때 실행이 된다. 따라서 for 루프를 사용할 때 함수 내부의 메시지가 출력이 된다.

```python
# Generator ex3
# filterfalse, accumulate, chain, product, groupby
import itertools

gen1 = itertools.count(1, 2.5)
```

```python
print(next(gen1))
print(next(gen1))
print(next(gen1))
print(next(gen1))
### ...무한
```

```python
# 조건
gen2 = itertools.takewhile(lambda n : n < 1000, itertools.count(1,2.5))

for v in gen2:
    print(v)
```

```python
# 필터 반대
gen3 = itertools.filterfalse(lambda n : n < 3, [1,2,3,4,5])

for v in gen3:
    print(v)
```

```python
# 누적 합계
gen4 = itertools.accumulate([x for x in range(1, 101)])

for v in gen4:
    print(v)
```

```python
# 연결1
gen5 = itertools.chain('ABCDE', range(1,11,2))

print(list(gen5))
```

```python
# 연결2

gen6 = itertools.chain(enumerate('ABCDE'))

print(list(gen6))
```

```python
# 개별
gen7 = itertools.product('ABCDE')

print(list(gen7))
```

```python
# 연산(경우의 수)
gen8 = itertools.product('ABCDE', repeat=2)

print(list(gen8))
```

```python
# 그룹화
gen9 = itertools.groupby('AAABBCCCCDDEEE')

print(list(gen9))
```

```python
for chr, group in gen9:
    print(chr, ' : ', list(group))
```
